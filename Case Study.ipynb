```python
import numpy as np
import pandas as pd
import seaborn as sb
import statistics as ss
```

```python
st=pd.read_csv("SP.csv")
st
```

```python
#Find out how many males and females participated in the test.

cg = st['gender'].value_counts()
cg
```

```python
#What do you think about the students' parental level of education?
pe=st['parental level of education'].value_counts(ascending=False)
print('Most of the parents are graduate \n\n),pe)
```

```python
#Who scores the most on average for math, reading and writing based on gender
st.groupby('gender').agg(
    mean_math_score=('math score','mean'),
    mean_reading_score=('reading score','mean'),
    mean_writing_score=('writing score','mean')
)
```

```python
#Who scores the most on average for math, reading and writing based on test preparation course
st.groupby('test preparation course').agg(
    mean_math_score=('math score','mean'),
    mean_reading_score=('reading score','mean'),
    mean_writing_score=('writing score','mean')
)
```

```python
#What do you think about the scoring variation for math, reading and writing
based on gender
pe = st[['gender','math score','reading score','writing score']].groupby('gender').std()
pe
```

```python
What do you think about the scoring variation for math, reading and writing
based on test preparation course
pe1 = st[['test preparation course','math score','reading score','writing score']].groupby('test preparation course').std()
pe1
```
